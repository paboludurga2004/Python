```python
def myFunc(*args,**kwargs):
    print("Non-Keyword Arguments (*args):")
    for arg in args:
        print(arg)
    print("\nKeyword Arguments (**kwargs):")
    for key,value in kwargs.items():
        print(f"{key}=={value}")
myFunc('Hey','Welcome',first='Geeks',mid='for',last='Geeks')
        
```

```python
#Recursive function
def factorial(n):
    if (n==0):
        return 1
    else:
        return n*(n-1)
print(factorial(5))
```
